```r
index <- Day - 1
```

```r
index
```

Daily Count

```r
#linear model
linear.model_dc <-lm(Daily.Count ~ index)
```

```r
summary(linear.model_dc)
```

```r
plot(index, Daily.Count, pch=20, xlab = "Day",ylab = "Daily.Count", cex.lab = 1.3, col = "#2eb8b3" )

abline(lm(Daily.Count ~ index), col = "dark green", lwd = 3)
grid()
```

Quadratic model

```r
index2 <- index^2
quadratic.model_dc <-lm(Daily.Count ~ index + index2)
```

```r
summary(quadratic.model_dc)
```

```r
boxplot(quadratic.model_dc[['residuals']],main='Boxplot: Residuals',ylab='residual value')
```

```r
timevalues <- seq(0, 206, 1)
predictedcounts <- predict(quadratic.model_dc,list(index=timevalues, index2=timevalues^2))
plot(index, Daily.Count, pch=20, xlab = "Day", ylab = "Daily.Count", cex.lab = 1.3, col = "#2eb8b3")
lines(timevalues, predictedcounts, col = "#a11b6e", lwd = 3)
grid()
```

Polynomial Model

```r
#Polynomial
index2 <- index^2
index3 <- index^3
poly.model_dc <-lm(Daily.Count ~ index + index2 + index3)
```

```r
summary(poly.model_dc)
```

```r
timevalues <- seq(0, 206, 1)
predictedcounts <- predict(poly.model_dc,list(index=timevalues, index2=timevalues^2, index3=timevalues^3))
plot(index, Daily.Count, pch=20, xlab = "Day", ylab = "Daily.Count", cex.lab = 1.3, col = "#2eb8b3")
lines(timevalues, predictedcounts, col = "red", lwd = 3)
grid()
```

Exponential Model

```r
#exponential
logdc <- log(Daily.Count)
logdc[which(!is.finite(logdc))] <- 0
```

```r
exponential.model_dc <- lm(logdc~ index)
summary(exponential.model_dc)
```

```r
timevalues <- seq(0, 206, 1)
Counts.exponential2 <- exp(predict(exponential.model_dc,list(Time=timevalues)))
plot(index, Daily.Count,pch=20, col = "#2eb8b3", xlab = "Day", ylab = "Daily.Count", cex.lab = 1.3 )
lines(timevalues, Counts.exponential2,lwd=3, col = "#3b51e3", xlab = "Time (s)", ylab = "Daily.Count")
grid()
```

```r
linear.model_1 <-lm(dd_by_dc ~ indexdc)
linear.model_2 <-lm(dd_by_dr ~ indexdr)
linear.model_3 <-lm(dd_by_da ~ indexda)
```

```r
print('dd_by_dc')
summary(linear.model_1)
print('dd_by_dr')
summary(linear.model_2)
print('dd_by_da')
summary(linear.model_3)
```

```r
options(repr.plot.width=18, repr.plot.height=6)
par(mfcol=c(1,3))
plot(indexdc, dd_by_dc,pch=16, col="#3b51e3", xlab = "Day", ylab = "dd_by_dc", cex.lab = 1.3, main="Deaths by Count")
abline(lm(dd_by_dc ~ indexdc), col = "red", lwd=3)
grid()
plot(indexdr, dd_by_dr,pch=16, col="#5dcc4b", xlab = "Day", ylab = "dd_by_dr", cex.lab = 1.3, main="Deaths by Recoveries")
abline(lm(dd_by_dr ~ indexdr), col = "red", lwd=3)
grid()
plot(indexda, dd_by_da,pch=16, col="#f26fac", xlab = "Day", ylab = "dd_by_da", cex.lab = 1.3, main="Deaths by Active")
abline(lm(dd_by_da ~ indexda), col = "red", lwd=3)
grid()
```

```r
indexdc2 <- indexdc^2
indexdr2 <- indexdr^2
indexda2 <- indexda^2
quadratic.model_1 <-lm(dd_by_dc ~ indexdc + indexdc2)
quadratic.model_2 <-lm(dd_by_dr ~ indexdr + indexdr2)
quadratic.model_3 <-lm(dd_by_da ~ indexda + indexda2)
```

```r
print('dd_by_dc')
summary(quadratic.model_1)
print('dd_by_dr')
summary(quadratic.model_2)
print('dd_by_da')
summary(quadratic.model_3)
```

```r
options(repr.plot.width=18, repr.plot.height=6)
par(mfcol=c(1,3))
timevalues <- seq(0, 206, 1)
predictedcounts1 <- predict(quadratic.model_1,list(index=timevalues, index2=timevalues^2))
plot(index, dd_by_dc,pch=16, col="#3b51e3", xlab = "Day", ylab = "dd_by_dc", cex.lab = 1.3, main="Deaths by Count")
lines(timevalues, predictedcounts1, col = "red", lwd = 3)
grid()
predictedcounts2 <- predict(quadratic.model_2,list(index=timevalues, index2=timevalues^2))
plot(index, dd_by_dr,pch=16, col="#5dcc4b", xlab = "Day", ylab = "dd_by_dr", cex.lab = 1.3, main="Deaths by Recovered")
lines(timevalues, predictedcounts2, col = "red", lwd = 3)
grid()
predictedcounts3 <- predict(quadratic.model_3,list(index=timevalues, index2=timevalues^2))
plot(index, dd_by_da,pch=16, col="#f26fac", xlab = "Day", ylab = "dd_by_da", cex.lab = 1.3, main="Deaths by Active")
lines(timevalues, predictedcounts3, col = "red", lwd = 3)
grid()
```
